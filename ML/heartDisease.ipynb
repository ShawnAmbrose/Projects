```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
#Reads the initial, unedited file and displays it

heart0=pd.read_csv(r"/content/drive/MyDrive/Colab_Notebooks/datasets/heart.csv")
print(heart0.to_string())
```

```python
#Reads the file with missing values

heart=pd.read_excel("/content/drive/MyDrive/Colab_Notebooks/datasets/heart1.xlsx")
print(heart.to_string())
```

```python
#Displays the X and Y sizes of the table in the csv or excel file

heart.shape
```

```python
#Displays the info of all the column values

heart.info()
```

```python
#Displays how many null values are in each column

heart.isnull().sum()
```

```python
#Used to drop all the rows with a null from the table

heart_dropped=heart.dropna()
heart_dropped.isnull().sum()
```

```python
#Displays table after dropping

print(heart_dropped.to_string())
```

```python
#Used to fill all the rows with a null from the table using mean, median, mode, etc...

heart_filled=heart
heart_filled.fillna(method="bfill",inplace=True)
heart_filled.isnull().sum()
```

```python
#Displays table after filling

print(heart_filled.to_string())
```

```python
#Used to show the first n values

heart_filled.head(50)
```

```python
#Used to show the last n values

heart_filled.tail(50)
```

```python
#Defining the input and outputs from the file
#Input => x
#Output => y

x=heart_filled[["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal"]]
y=heart_filled[["target"]]

#Displaying the values in x and y
print(x,y)
```

```python
#Spiliting the train values and the test values

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#Using svm algorithm to find the accuracy

from sklearn import svm
svm=svm.SVC()
svm.fit(x_train, y_train)
y_pred=svm.predict(x_test)
score_svm=accuracy_score(y_pred,y_test)
print(score_svm*100)
```

```python

#Using desition tree algorithm to find the accuracy

from sklearn import tree
dt=tree.DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred=dt.predict(x_test)
score_dt=accuracy_score(y_pred,y_test)
print(score_dt*100)
```

```python

#Using random forest algorithm to find the accuracy


from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10)
rf.fit(x_train, y_train)
y_pred=rf.predict(x_test)
score_rf=accuracy_score(y_pred,y_test)
print(score_rf*100)
```

```python
#Using xg boost algorithm to find the accuracy

import xgboost as xgb
xg = xgb.XGBClassifier(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,max_depth = 5, alpha = 10, n_estimators = 10)
xg.fit(x_train, y_train)
y_pred=xg.predict(x_test)
score_xg=accuracy_score(y_pred,y_test)
print(score_xg*100)
```
